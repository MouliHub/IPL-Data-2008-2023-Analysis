# Performance trend of teams over different seasons

```python
# Group the data by season and team, and calculate the number of matches won by each team in each season
team_season_wins = df.groupby(['season', 'winner'])['id'].count().reset_index()
team_season_wins.rename(columns={'id': 'Wins'}, inplace=True)

# Pivot the data to get the number of wins for each team in each season
team_season_pivot = team_season_wins.pivot(index='season', columns='winner', values='Wins').fillna(0)

# Plot the performance trend for each team
team_season_pivot.plot(kind='line', marker='o', figsize=(22, 10))
plt.title('Performance Trend of Teams Over Different Seasons')
plt.xlabel('season')
plt.ylabel('Number of Wins')
plt.legend(loc='upper left')
plt.xticks(team_season_pivot.index)
plt.grid(True)
plt.show()
```

# The top 4 teams in the entire IPL season

```python

# Filter the data for the seasons 2008-2023
filtered_df = df[(df['season'] >= 2008) & (df['season'] <= 2023)]

# Calculate the number of matches played by each team
matches_played = filtered_df['home_team'].value_counts().add(filtered_df['away_team'].value_counts(), fill_value=0).reset_index()
matches_played.columns = ['Team', 'Matches Played']

# Calculate the number of matches won by each team
matches_won = filtered_df['winner'].value_counts().reset_index()
matches_won.columns = ['Team', 'Matches Won']

# Merge the matches played and matches won dataframes
team_stats = pd.merge(matches_played, matches_won, on='Team', how='left')

# Calculate the win percentage for each team
team_stats['Win Percentage'] = (team_stats['Matches Won'] / team_stats['Matches Played']) * 100

# Sort the teams based on their matches won in descending order
team_stats = team_stats.sort_values('Matches Won', ascending=False).head(4)

# Print the team statistics
print("Team Statistics (2008-2023):")
print(team_stats[['Team', 'Matches Played', 'Matches Won', 'Win Percentage']].to_string(index=False))
```

# Top 10 Highest scores in IPL (2008-2023)

```python
# Filter the data for the years 2008-2023
df = df[(df['season'] >= 2008) & (df['season'] <= 2023)]

# Create a new DataFrame to store the highest scores for each team
team_scores = pd.DataFrame(columns=['Team', 'Highest Score'])

# Iterate over each team
for team in df['winner'].unique():
    # Filter the data for the current team
    team_data = df[df['winner'] == team]
    
    # Find the highest score in the first innings
    max_score_1st = team_data['1st_inning_score'].max()
    
    # Find the highest score in the second innings
    max_score_2nd = team_data['2nd_inning_score'].max()
    
    # Determine the highest score overall
    highest_score = max(max_score_1st, max_score_2nd)
    
    # Add the team and highest score to the DataFrame
    team_scores = team_scores.append({'Team': team, 'Highest Score': highest_score}, ignore_index=True)

# Sort the team scores in descending order
team_scores = team_scores.sort_values('Highest Score', ascending=False).head(10)

# Print the top 10 highest scores by team (Team, Highest Score)
print("Top 10 Highest Scores by Team in IPL (2008-2023):")
print(team_scores.to_string(index=False))
```

# Win Percentage of each team

```python
# Calculate win percentage
team_wins = df['winner'].value_counts()
team_matches = df['home_team'].value_counts() + df['away_team'].value_counts()
win_percentage = (team_wins / team_matches) * 100

# Create bar chart
plt.figure(figsize=(10, 6))
plt.bar(win_percentage.index, win_percentage.values)
plt.xlabel('Teams')
plt.ylabel('Win Percentage')
plt.title('Win Percentage of Each Team')
plt.xticks(rotation=45)
plt.show()
```

# Home win percentage of each team

```python
import seaborn as sns

# Calculate home win count
home_wins = df[df['winner'] == df['home_team']]['home_team'].value_counts()

# Calculate home match count
home_matches = df['home_team'].value_counts()

# Calculate home win percentage
home_win_percentage = (home_wins / home_matches) * 100

# Sort data in descending order
home_win_percentage = home_win_percentage.sort_values(ascending=False)

# Create bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=home_win_percentage.index, y=home_win_percentage.values)
plt.xlabel('Teams')
plt.ylabel('Home Win Percentage')
plt.title('Home Win Percentage of Each Team')
plt.xticks(rotation=45)
plt.show()
```

# Away win percentage of each team

```python
import seaborn as sns

# Calculate away win count
away_wins = df[df['winner'] == df['away_team']]['away_team'].value_counts()

# Calculate away match count
away_matches = df['away_team'].value_counts()

# Calculate away win percentage
away_win_percentage = (away_wins / away_matches) * 100

# Sort data in descending order
away_win_percentage = away_win_percentage.sort_values(ascending=False)

# Create bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=away_win_percentage.index, y=away_win_percentage.values)
plt.xlabel('Teams')
plt.ylabel('Away Win Percentage')
plt.title('Away Win Percentage of Each Team')
plt.xticks(rotation=45)
plt.show()
```

# Average runs per match by team

```python

# Filter the dataset for matches played from 2008 to 2023
df = df[df['season'].between(2008, 2023)]

# Calculate the total runs scored by each team in home matches
home_runs = df.groupby('home_team')['home_runs'].sum()

# Calculate the total runs scored by each team in away matches
away_runs = df.groupby('away_team')['away_runs'].sum()

# Calculate the total number of matches played by each team
matches_played = df['home_team'].value_counts() + df['away_team'].value_counts()

# Calculate the average runs scored per match by each team
avg_runs_per_match = (home_runs + away_runs) / matches_played

# Create a new DataFrame to store the average runs per match for each team
avg_runs = pd.DataFrame({'Team': avg_runs_per_match.index,
                         'Average Runs per Match': avg_runs_per_match.values})

# Sort the DataFrame by average runs per match in descending order
avg_runs = avg_runs.sort_values('Average Runs per Match', ascending=False)

# Plot the average runs per match for each team
plt.figure(figsize=(12, 6))
plt.bar(avg_runs['Team'], avg_runs['Average Runs per Match'], alpha=0.7)
plt.xticks(rotation=90)
plt.xlabel('Team')
plt.ylabel('Average Runs per Match')
plt.title('Average Runs per Match by Team (2008-2023)')
plt.show()
```

# Highest number of player of match awards in IPL 2008-2023

```python

# Count the number of times each player has won the Player of the Match (POM) award
pom_count = df['pom'].value_counts()

# Get the top 5 players with the highest count of POM awards
top_pom_players = pom_count.head(5)

# Plot the top 5 players
plt.figure(figsize=(10, 6))
top_pom_players.plot(kind='bar')
plt.title('Top 5 Players with Most Player of the Match (POM) Awards')
plt.xlabel('Player')
plt.ylabel('Number of POM Awards')
plt.xticks(rotation=45)
plt.show()
```

# Highest total by each team

```python
import matplotlib.pyplot as plt

# Calculate highest scores by each team
home_high_score = df.groupby('home_team')['1st_inning_score'].max()
away_high_score = df.groupby('away_team')['2nd_inning_score'].max()
highest_scores = pd.concat([home_high_score, away_high_score], axis=1).max(axis=1)

# Sort the highest scores in descending order
highest_scores = highest_scores.sort_values(ascending=False)

# Plot the highest scores
plt.figure(figsize=(12, 6))
plt.bar(highest_scores.index, highest_scores.values)
plt.xlabel('Team')
plt.ylabel('Highest Score')
plt.title('Highest Total Score by Each Team')
plt.xticks(rotation=90)
plt.show()
```
